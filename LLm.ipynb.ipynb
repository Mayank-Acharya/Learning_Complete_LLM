Token level embeddings


```python
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-v3-xsmall")   # ← note: deberta_base doesn't exist
model = AutoModel.from_pretrained("microsoft/deberta-v3-xsmall")

text = "Hii My Name IS MAYANK KUMAR ACHARYA And I Am Learning Embeddings which are just vectoRs"

tokens = tokenizer(text, return_tensors="pt")

print(tokens)
```

```python
embeddings = model(**tokens)[0]
```

```python
embeddings
```

```python
embeddings.shape
```

**Sentence Level Embeddings**


```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-mpnet-base-v2")
embeddings = model.encode("Hii My Name IS MAYANK KUMAR ACHARYA And I Am Learning Embeddings which are just vectoRs")
```
